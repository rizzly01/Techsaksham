```python
# Group data by Season and Crop, summing Production
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('IAS.csv')
season_crop_production = df.groupby(['Season', 'Crop'])['Production'].sum().reset_index()

# Sort data for better visualization
season_crop_production = season_crop_production.sort_values(by='Production', ascending=False)

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(data=season_crop_production, x='Season', y='Production', hue='Crop', palette='tab20')

plt.title('Total Crop Production by Season and Crop')
plt.xlabel('Season')
plt.ylabel('Total Production')
plt.legend(title='Crop', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#crop to be produce as per season wise  
# Get the top crop by production for each season
top_crop_per_season = season_crop_production.loc[season_crop_production.groupby('Season')['Production'].idxmax()]

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_crop_per_season, x='Season', y='Production', hue='Crop', palette='pastel')

plt.title('Top Crop by Production for Each Season')
plt.xlabel('Season')
plt.ylabel('Total Production')
plt.legend(title='Top Crop', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Show the top crop for each season in text format
top_crop_per_season[['Season', 'Crop', 'Production']]
```

```python
# Group data by year and crop, summing production
year_crop_production = df.groupby(['Crop_Year', 'Crop'])['Production'].sum().reset_index()

# Find the top-producing crop for each year
top_crop_per_year = year_crop_production.loc[year_crop_production.groupby('Crop_Year')['Production'].idxmax()]

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(data=top_crop_per_year, x='Crop_Year', y='Production', hue='Crop', palette='tab10')

plt.title('Top Crop by Production Each Year')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.legend(title='Crop', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display the top crop for each year
top_crop_per_year[['Crop_Year', 'Crop', 'Production']]
```

```python
# realtion between area and production

import scipy.stats as stats

# Scatter plot to visualize the relationship between Area and Production
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Area', y='Production', alpha=0.5, color='blue')

plt.title('Relationship Between Area and Production')
plt.xlabel('Area (in hectares)')
plt.ylabel('Production (in tons)')
plt.show()

# Calculate correlation coefficient
correlation = df[['Area', 'Production']].corr().loc['Area', 'Production']
correlation
print("relation between area and productionThe relationship between Area and Production in your dataset shows a very weak positive")
print("correlation — with a correlation coefficient of about 0.041.")
```

```python
# Group data by State and sum the Production
state_production = df.groupby('State_Name')['Production'].sum().reset_index()

# Sort in descending order and get top 10 states
top_states = state_production.sort_values(by='Production', ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=top_states, x='State_Name', y='Production', palette='Blues_d')

plt.title('Top 10 States by Total Crop Production')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display top states
top_states
```

```python
# Group data by State and Crop, summing Production
state_crop_production = df.groupby(['State_Name', 'Crop'])['Production'].sum().reset_index()

# Find the top crop for each state by production
top_crop_per_state = state_crop_production.loc[state_crop_production.groupby('State_Name')['Production'].idxmax()]

# Sort by production for better visualization
top_crop_per_state = top_crop_per_state.sort_values(by='Production', ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(data=top_crop_per_state, x='State_Name', y='Production', hue='Crop', palette='Paired')

plt.title('Top Crop by Production in Each State (Top 10)')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.legend(title='Crop', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Show top crop for each state
top_crop_per_state[['State_Name', 'Crop', 'Production']]
```
